```python
from nltk import *
import random
import matplotlib
%matplotlib inline
```

```python
with open("../books/The-Hobbit-by-J.R.R-Tolkien.txt") as f:
    words = f.read().split()
words = Text(words)
```

```python
length_words = len(words)
uniq_words = set(words)
print(length_words)
print(len(uniq_words))
print(len(uniq_words) / length_words)
```

```python
" ".join([words[random.randint(0,length_words-1)] for i in range(250)])
```

```python
fdist = FreqDist(words)
fdist.plot(50, cumulative=False)
```

```python
" ".join(sorted(w for w in uniq_words if len(w) > 7 and fdist[w] > 15))
```
